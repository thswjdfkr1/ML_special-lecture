```python
from sklearn.preprocessing import StandardScaler

ss_col = X.select_dtypes(exclude='object').columns

ss = StandardScaler()
X[ss_col] = ss.fit_transform(X[ss_col])
```

```python
obj_col = X.select_dtypes(include=['object']).columns

for col in obj_col:
  print(col)
  print(X[col].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

for col in obj_col:
  le = LabelEncoder()
  X[col] = le.fit_transform(X[col])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,random_state=810)
train_x, val_x, train_y, val_y = train_test_split(x_train, y_train, test_size=0.2, stratify=y_train, random_state=810)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(10, ), activation='logistic',
                      solver='sgd', alpha=0.01, batch_size=32, learning_rate_init=0.1, max_iter=1000,
                      random_state=810)

model = MLPClassifier(random_state=810)

model.fit(train_x, train_y)
```

```python
from sklearn.metrics import classification_report, accuracy_score

pred_val = model.predict(val_x)
pred_test = model.predict(x_test)

print('val:', model.score(val_x, val_y))
print('test:', model.score(x_test, y_test))
# print('val:', classification_report(val_y, pred_val))
# print('val:', accuracy_score(val_y, pred_val))
# print('test:', classification_report(y_test, pred_test))
# print('test:', accuracy_score(y_test, pred_test))
```
